## EDA

```python
df.columns
```

## Gender

```python
ax = sns.countplot(x = 'Gender', data = df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum()
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

From Above the most of the buyer are Female 
And Purchaing Power of Female buyers are more than Man
Females have spent more Amount on buying rather than Male

## Age Group

```python
age_g = sns.countplot(data=df, x='Age Group')
```

from this in insights its clear that there were 
more number of peoples of age (26-35) in this data set

```python
age_g = sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in age_g.containers:
    age_g.bar_label(bars)
```

Females of Age Group 26 to 35 are more in the Shopping

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum()
sales_age
```

```python
sns.barplot(x='Age Group', y='Amount', data= sales_age)
```

from above graph we can see that most of the buyers are of age between 26-35 yrs 

```python
df.columns
```

## Sales

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
```

```python
sales_state
```

```python
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data= sales_state, x='State' ,y='Orders')
```

from this graph we analyse that Uttar Pradesh State has maximum orders
and Telengana had lowest orders

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(data= sales_state, x='State' ,y='Amount')
```

from this graph we analyse that Uttar Pradesh  had spent highest Amount on shopping
and Telengana had spent Lowest A,ount on Shopping

## Marital Status

```python
sns.set(rc={'figure.figsize':(6,5)})
marital_g = sns.countplot(x='Marital_Status', data = df)

for i in marital_g.containers:
    marital_g.bar_label(i)
```

```python
sales_marr = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_marr
```

```python
sns.barplot(x="Marital_Status", y="Amount", hue='Gender', data = sales_marr)
```

from the above graph we can see that most of the buyers are married womens

```python
sns.set(rc={'figure.figsize':(20,5)})
occ = sns.countplot(data= df, x='Occupation')
for i in occ.containers:
    occ.bar_label(i)
```

## Product_Categorey

```python
prod_c = sns.countplot(data=df, x="Product_Category")
sns.set(rc={'figure.figsize':(30,12)})
for i in prod_c.containers:
    prod_c.bar_label(i)
```

from this graph we can say that Clothing & apperal has more ordered and second is food

```python
sales_prod = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(8)
sns.barplot(x='Product_Category',y='Amount',data=sales_prod)
sns.set(rc={'figure.figsize':(16,12)})
```

from this graph we can say that Food has more sold and second is Clothing
